Each row in the assignment datafile corresponds to a single observation.

The following variables are provided to you:

* **id** : station identification code
* **date** : date in YYYY-MM-DD format (e.g. 2012-01-24 = January 24, 2012)
* **element** : indicator of element type
    * TMAX : Maximum temperature (tenths of degrees C)
    * TMIN : Minimum temperature (tenths of degrees C)
* **value** : data value for element (tenths of degrees C)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Importing data set
df = pd.read_csv('data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')
df.head()
```

```python
#Creating columns month, day, year
df['Date'] = pd.to_datetime(df['Date'])
df['month'] = pd.DatetimeIndex(df['Date']).month
df['day'] = pd.DatetimeIndex(df['Date']).day
df['year'] = pd.DatetimeIndex(df['Date']).year

#Multipling Data_Value per 0.1 to convert temperature for (ªC) Celsius
df['Data_Value'] = df['Data_Value']*0.1

df.set_index(['month', 'day'], inplace=True)
df.head()
```

```python
#Removing day 02-29 because it may only 365 per year
selected_df = df.loc[2,29] 
df = df[~df.index.isin(selected_df.index)]

#Two data sets were created for the period 2004-2015 and another for only 2015
df_period = df[df['year'] != 2015]
df_2015 = df[df['year'] == 2015]

#Creating four data frame - two for Temperatures Max and two for Temperatures Min 
temp_max_df = df_period[df_period['Element'] == 'TMAX']
temp_min_df = df_period[df_period['Element'] == 'TMIN']

temp_max_df_2015 = df_2015[df_2015['Element'] == 'TMAX']
temp_min_df_2015 = df_2015[df_2015['Element'] == 'TMIN']

#Take the records temperatures for 365 days (1 year) - per period of 2005-2014
temp_max = temp_max_df.groupby(level = ['month', 'day']).max()[['Data_Value', 'Date']]
temp_min = temp_min_df.groupby(level = ['month', 'day']).min()[['Data_Value', 'Date']]

#Take the records temperatures for 365 days (1 year) - only 2015
temp_max_2015 = temp_max_df_2015.groupby(level = ['month', 'day']).max()[['Data_Value', 'Date']]
temp_min_2015 = temp_min_df_2015.groupby(level = ['month', 'day']).min()[['Data_Value', 'Date']]
```

```python
%matplotlib notebook
import matplotlib.dates as mdates
import matplotlib.ticker as ticker

months = mdates.MonthLocator()
monthsFmt = mdates.DateFormatter('%m')

#Creating a data array for plotting range with 365 days
date_range = temp_max['Date'].unique() 

#Creating a figure and an axis
fig, ax = plt.subplots(figsize=(10, 5))

#Plotting a graph
ax.plot(date_range , list(temp_max['Data_Value']), color='#ec8e49', label='2005-2014 High')
ax.plot(date_range ,list(temp_min['Data_Value']), color='#5abfe7', label='2005-2014 Low')

#Formating the ticks of x-axis
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_minor_locator(mdates.MonthLocator(bymonthday=15))
ax.xaxis.set_major_formatter(ticker.NullFormatter())
ax.xaxis.set_minor_formatter(mdates.DateFormatter('%b'))

#Setting datemin and datemax in the x-axis with only 365
datemin = np.datetime64(temp_max['Date'][0], 'Y')
datemax = np.datetime64(temp_max['Date'][-1], 'Y') + np.timedelta64(1, 'Y')
ax.set_xlim(datemin, datemax)


#Setting the labels ticks in the middle of the months
for tick in ax.xaxis.get_minor_ticks():
    tick.tick1line.set_markersize(0)
    tick.tick2line.set_markersize(0)
    tick.label1.set_horizontalalignment('center')

#Filling between the region of twos lines plotted
ax.fill_between(date_range, temp_min['Data_Value'], temp_max['Data_Value'], color='Blue', alpha=0.1)
plt.show()
```

```python
#Setting temperatures in the y-axis with -40º and 50º
plt.ylim(-40, 50)

#Removing the first label tick(-40ª) and the last (50) to make the graph looks clear
locs, lables = plt.yticks()
lables[0].set_visible(False)
lables[-1].set_visible(False)

#Setting a label for the y-axis
plt.ylabel('Temperatures $(^{\circ}$C)')
```

```python
#Comparing the dates of 2015 and setting the dates that are higher or lower than period of 2005-2014 with True
temp_max_2015['Higher'] = temp_max_2015['Data_Value'] > temp_max['Data_Value']
temp_max_2015[temp_max_2015['Higher'] == False] = np.nan
temp_min_2015['Lower'] = temp_min_2015['Data_Value'] < temp_min['Data_Value']
temp_min_2015[temp_min_2015['Lower'] == False] = np.nan

#Scattering temperatures that are higher and lower
plt.scatter(date_range, temp_min_2015['Data_Value'], marker='o', s=15, color='#1c91c0', label='2015 Lower')
plt.scatter(date_range, temp_max_2015['Data_Value'], marker='o', s=15, color='#e7711b', label='2015 Higher')

#Creating title for graph
plt.title('Ann Arbor Records Temperature (2005-2014) and 2015')

#Creating legend and removing the frame
plt.legend(loc=0, frameon=False)

#To make the graph fit in the figure
fig.tight_layout()
```

```python
#%matplot notebook
#Saving the figure
fig.savefig('plot.png', bbox_inches='tight')   
plt.close(fig)
```
